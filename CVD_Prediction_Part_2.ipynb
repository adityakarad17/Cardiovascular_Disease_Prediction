## **Importing libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **Importing dataset**

```python
df = pd.read_csv("cardio_train.csv", sep = ";")
df.head(10)
```

```python
import seaborn as sns
sns.lmplot('height', 'weight', df, hue='cardio', fit_reg=False)
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.show()
```

```python
df['cholesterol'].value_counts().plot(kind='barh')
```

```python
df['smoke'].value_counts().plot(kind='barh')
```

```python
df['cardio'].value_counts().plot(kind='barh')
```

```python
import numpy as np    
idx_1 = np.where(df.cardio == 1)
idx_0 = np.where(df.cardio == 0)
```

```python
plt.scatter(df.iloc[idx_1].cholesterol, df.iloc[idx_1].weight, s=10, c='b', marker="o", label='Present')
plt.scatter(df.iloc[idx_0].cholesterol, df.iloc[idx_0].weight, s=10, c='r', marker="o", label='Absent')
plt.ylabel('Weight')
plt.xlabel('Cholesterol')
plt.legend(loc="upper right")
plt.title('Weight VS Cholesterol')
plt.show()
```

```python
plt.scatter(df.iloc[idx_1].gluc, df.iloc[idx_1].weight, s=10, c='b', marker="o", label='Present')
plt.scatter(df.iloc[idx_0].gluc, df.iloc[idx_0].weight, s=10, c='r', marker="o", label='Absent')
plt.ylabel('Weight')
plt.xlabel('Glucose Intake')
plt.legend(loc="upper right")
plt.title('Weight VS Glucose Intake')
plt.show()
```

```python
plt.scatter(df.iloc[idx_1].ap_lo, df.iloc[idx_1].weight, s=0.7, c='b', marker="o", label='Present')
plt.scatter(df.iloc[idx_0].ap_lo, df.iloc[idx_0].weight, s=0.7, c='r', marker="o", label='Absent')
plt.ylabel('Weight')
plt.xlabel('Diastolic blood pressure')
plt.legend(loc="upper right")
plt.xlim(0,400)
plt.title('Weight VS Diastolic blood pressure')
plt.show()
```

```python
plt.scatter(df.iloc[idx_1].ap_hi, df.iloc[idx_1].weight, s=2, c='b', marker="o", label='Present')
plt.scatter(df.iloc[idx_0].ap_hi, df.iloc[idx_0].weight, s=2, c='r', marker="o", label='Absent')
plt.ylabel('Weight')
plt.xlabel('Systolic blood pressure')
plt.legend(loc="upper right")
plt.xlim(0,400)
plt.title('Weight VS Systolic blood pressure')
plt.show()
```

```python
df.hist(bins=15, color='red', edgecolor='black', linewidth=1.0,
           xlabelsize=8, ylabelsize=8, grid=False)    
plt.tight_layout(rect=(0, 0, 1.2, 1.2)) 
```

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10, 6))
corr = df.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Pathological Data Attributes Correlation Heatmap', fontsize=14)
```

```python
# Pair-wise Scatter Plots
cols = ['cardio','height','weight','ap_lo','ap_hi']
pp = sns.pairplot(df[cols], size=1.8, aspect=1.8,
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Pathological Data Attributes Pairwise Plots', fontsize=14)
```
